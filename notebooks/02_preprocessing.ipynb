```python
import pandas as pd

df_eq=pd.read_csv('data/raw/earthquake.csv')
df_eq.head()
```

```python
df_eq = df_eq[['time', 'mag', 'place']]
```

```python
df_eq = df_eq.rename(columns={
    'time': 'date',
    'mag': 'magnitude'
})
```

```python
df_eq['date'] = pd.to_datetime(df_eq['date'])
```

```python
df_eq = df_eq[df_eq['magnitude'] >= 5]
```

```python
df_eq = df_eq.sort_values(by='date')
```

```python
df_eq.to_csv('results/earthquakes_clean.csv', index=False)
```
