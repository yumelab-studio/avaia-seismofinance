```python
import yfinance as yf
import pandas as pd
stock = yf.download("8766.T", start="2010-01-01")
stock.head()
```

```python
stock.to_csv('../data/raw/stocks.csv')
```

```python
stock = stock[['Close']]
stock.reset_index(inplace=True)
```

```python
stock.rename(columns={
    'Date': 'date',
    'Close': 'close'
}, inplace=True)
```

```python
stock['date'] = pd.to_datetime(stock['date'])
```

```python
stock['return'] = stock['close'].pct_change()
```

```python
stock = stock.sort_values('date')
```

```python
stock.to_csv('../data/processed/stocks_clean.csv', index=False)
```
